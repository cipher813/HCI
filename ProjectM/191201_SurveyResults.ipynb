```python
import pandas as pd
```

```python
wizard = pd.read_csv("wizard.csv")
template = pd.read_csv("template.csv")
```

```python
wizard.head()
```

```python
def question_result(dataframe, column):
    df = dataframe.groupby(column).agg({"response":"count"})
    df.rename(columns={"response":"count"},inplace=True)
    df['percent'] = 100 * (df / df.sum())
    return df.round(1)

def prototype_result(dataframe,n_questions):
    for i in range(1,n_questions + 1):
        print(question_result(wizard,f"Q{i}"))
        print("\n")
```

```python
prototype_result(wizard,14)
```

```python
prototype_result(template,14)
```
